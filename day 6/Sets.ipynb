# Set
A set is a collection which is unordered and unindexed. In Python sets are written with curly brackets.

## Example
Create a set:

```python
a = {'apple','mango','banana'}
print(a)
```

sets are unordered, so we cannot be sure in which order the items will appear.

## Access items
We cannot access items in a set by referring an index, since sets are unordered the items has no index.
But we can loop through the set items by using a 'for' loop and 'in' keyword

```python
for i in a:
    print(i)
```

## Example
Check if 'banana' is present in this set

```python
print('banana' in a)
```

## Change items
Once a set is created, we cannot change its items, but we can add new items.

## Add items:
To add one item to a set, we wil use 'add()' method.
To add more than one items to a set we use the 'update()' method

## Example
Add an item to a using add() method

```python
a.add('orange')
print(a)
```

## Example
Add multiple items to 'a' using update() method

```python
a.update(['strawberry','orange','chickoo','watermelon'])
print(a)
```

## Get the length of a set
To determine how many items a set has ,use the 'len()' method

## Example
Get the number of items in 'a' set

```python
print(len(a))
```

## Remove item
To remove an item in a set, use the 'remove()' or the 'discard()' method

```python
a.remove('mango')
a.discard('watermelon')
print(a)
```

However, if that item doesn't exist in the given set, remove() method will raise an error and discard method will not raise an error

```python
a.remove('grapes')
```

```python
a.discard('grapes')
```

### pop() will remove the last item and returns it to the function

However, since sets are unordered, we cannot say which item will get removed

```python
a = {'apple','mango','chickoo','banana','grapes'}
b = a.pop()
print(a,'\n \n \n', b)
```

## Example
The 'clear()' method will empty the sets

```python
a = {'apple','mango','chickoo','banana','grapes'}
a.clear()
print(a)
```

## Example
The 'del' keyword will delete the set completely

```python
a = {'apple','mango','chickoo','banana','grapes'}
del a
print(a)
```

## Join 2 sets
the 'union(). method will join 2 sets a return a new set with combined items from both the sets

```python
a = {1,2,3,4,5,6}
b = {7,8,9,10,11,12,13}
c = a.union(b)
print(c)
```

## Example
The'update()' method will insert the items in set2 into set1

```python
a = {1,2,3,4,5,6}
b = {7,8,9,10,11,12,13}
a.update(b)
print(a)
```

#### Both union and update methods will exclude duplicate items

## The set constructor
use the 'set()' method to make a set

```python
a = set(('apple','mango','banana'))
# double round brackets will be used
print(a)
```
