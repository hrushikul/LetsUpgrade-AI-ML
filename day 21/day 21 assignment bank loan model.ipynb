# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# importing the dataset 

```python
dataset = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name= 1)
dataset.head()
```

# checking for null values

```python
dataset.isna()
```

```python
dataset.describe()
```

```python
dataset.isna().sum().sum()
```

No null values

# dividing dependent and independent variables

```python
dataset = dataset.drop(columns=['ID','ZIP Code'])
dataset.head(2)
```

```python
X = dataset.iloc[:,[0,1,2,3,4,5,6,8,9,10,11]].values
y = dataset.iloc[:,7].values
```

# Splitting the dataset into test set and training set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Fitting the model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

# Predicting the results

```python
print(classifier.predict(sc.transform([[38,14,130,4,4.70,3,134,0,0,0,0]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred) 
print(cm)
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print(tn, fp, fn, tp)
accuracy_score(y_test, y_pred)
```

#### The model is now created for bank loan prediction with accuracy score of 95%
